# Get Facebook stock prices for the last three years

```python
start = date(2016,3,21)
end = date(2019,3,21)
ticker = 'FB'
data_source = 'iex'

facebook_prices = DataReader(ticker, data_source, start, end)
```

# Plot the stock prices

```python
facebook_prices['close'].plot(title=ticker)
plt.show()
```

# Display Economic Data form the Federal Reserve

```python
start = date(2016,3,21)
# skip end date to obtain the most recent data
series_code = 'DGS10'
data_source = 'fred'

ten_year_treasury_rate = DataReader(series_code, data_source, start)
```

```python
ten_year_treasury_rate.plot(title='10 Year Treasory Rate')
plt.show()
```

```python
# Get oil prices with DCOILWTICO
start = date(2014,3,21)
# skip end date to obtain the most recent data
series_code = 'DCOILWTICO'
data_source = 'fred'

oil = DataReader(series_code, data_source, start)
```

```python
oil.plot(title='10 Year Treasory Rate')
plt.show()
```

# Compare Bond and Stock Performance

```python
start = date(2018,1,1)

series = ['BAMLHYH0A0HYM2TRIV', 'SP500']

# Import the data
data = DataReader(series, 'fred', start)

# Plot the results
data.plot(subplots=True, title='Performance Comparison')

# Show the plot
plt.show()
```
