# Get Facebook stock prices for the last three years

```python
start = date(2016,3,21)
end = date(2019,3,21)
ticker = 'FB'
data_source = 'iex'

facebook_prices = DataReader(ticker, data_source, start, end)
```

# Plot the stock prices

```python
facebook_prices['close'].plot(title=ticker)
plt.show()
```

# Display multiple stocks

```python
tickers = ['AAPL', 'FB', 'AMZN', 'NFLX', 'MSFT']
start = date (2016,3,22)

data = DataReader(tickers, 'iex', start)

print(type(data))
data.info()
data.head()
```

```python
close_data = data['close']
close_data.head(10)
```

```python
close_data.plot(subplots=True)
plt.tight_layout()
plt.show()
```

## Calculating quantiles

```python
some_quantile = close_data.quantile(0.5)
median = close_data.median()
median == some_quantile # of course the median is equal to the 50% quantile
```

```python
quartiles = close_data['AAPL'].quantile([.25, .75])
print(quartiles)
print("APPL inter-quartile range: {}".format(quartiles[.75] - quartiles[.25]))
```

```python
import numpy as np

deciles = np.arange(start=0.1, stop=0.91, step=0.1)
deciles = close_data['AAPL'].quantile(deciles)
print(deciles)
```

```python
deciles.plot(kind='bar', title="AAPL Deciles")
plt.show()
```
