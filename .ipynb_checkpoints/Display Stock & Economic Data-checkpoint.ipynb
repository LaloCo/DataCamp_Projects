# Get Facebook stock prices for the last three years

```python
start = date(2016,3,21)
end = date(2019,3,21)
ticker = 'FB'
data_source = 'iex'

facebook_prices = DataReader(ticker, data_source, start, end)
```

# Plot the stock prices

```python
facebook_prices['close'].plot(title=ticker)
plt.show()
```
